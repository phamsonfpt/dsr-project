# Khám phá dữ liệu Job Market bằng SQL

Notebook này được tạo ra để bạn có thể dễ dàng viết các câu truy vấn SQL (Query) và xem kết quả dưới dạng bảng cực kỳ trực quan nhờ thư viện `pandas`.

```python
import sqlite3
import pandas as pd

# Mở kết nối đến file cơ sở dữ liệu SQLite
conn = sqlite3.connect('database/job_market.sqlite')
```

### 1. Xem danh sách công việc (Bảng `jobs_clean`)

```python
query_jobs = """
SELECT job_id, title, company, salary_min, salary_max, location, source 
FROM jobs_clean 
ORDER BY scraped_at DESC 
LIMIT 10;
"""

# Đọc dữ liệu từ SQL và hiển thị thành bảng
df_jobs = pd.read_sql_query(query_jobs, conn)
df_jobs
```

### 2. Xem các kỹ năng yêu cầu cho từng công việc (Bảng `job_skills`)

```python
query_skills = """
SELECT j.title, j.company, s.skill_name
FROM jobs_clean j
JOIN job_skills s ON j.job_id = s.job_id
WHERE s.skill_name LIKE '%python%' -- Thay đổi từ khóa ở đây
LIMIT 10;
"""
df_skills = pd.read_sql_query(query_skills, conn)
df_skills
```

### 3. Xem báo cáo xu hướng thị trường (Bảng `market_trends`)

```python
query_trends = """
SELECT category, key_name, value_metric, extra_json
FROM market_trends
"""
df_trends = pd.read_sql_query(query_trends, conn)
df_trends
```

### 4. Truy vấn tự do của bạn
Bạn có thể viết bất kỳ lệnh `SELECT` nào bạn muốn ở dưới đây:

```python
my_query = """
-- Viết SQL của bạn tại đây
SELECT * 
FROM jobs_clean 
WHERE company LIKE '%fpt%'
"""
df_custom = pd.read_sql_query(my_query, conn)
df_custom
```
